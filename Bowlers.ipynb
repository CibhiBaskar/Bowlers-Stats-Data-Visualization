```python
import pandas as pd
import numpy as np

dt = pd.read_csv("C:/Users/User/Desktop/Data Sets/Bowler_data.csv")
df = pd.DataFrame(dt)
df.head(20)
```

```python
df = df.replace('-',np.nan)
```

```python
df.drop(columns = ['Unnamed: 0'], axis = 1)
```

```python
df = df.replace(np.nan,0)
```

```python
df = df[['Overs','Mdns','Runs','Wkts','Econ','Ave','Bowler']]
```

```python
df['Overs'] = df['Overs'].astype(np.float)
```

```python
df = df.astype({'Mdns':'int64'})
```

```python
df = df.astype({'Runs':'int64'})
```

```python
df = df.astype({'Wkts':'int64'})
```

```python
df = df.astype({'Econ':'float'})
```

```python
df = df.astype({'Ave':'float'})
```

```python
df.dtypes
```

```python
df = df.replace(0,np.nan)
```

```python
df.dropna(subset = ['Overs'], axis = 0, inplace=True)
```

```python
df = df.groupby(['Bowler']).sum()
df
```

```python
df = df.reset_index()
df
```

```python
df['Ave'] = df['Runs']/df['Wkts']
```

```python
df['Econ'] = df['Runs']/df['Overs']
```

```python
df.drop(columns = ['Overs'])
```

```python
df = df[df['Wkts'] >= 10.0]
df = df[df['Ave'] <= 25.0]
df
```

```python
import matplotlib.pyplot as mp
import seaborn as sns
```

```python
width = 22
height = 6
mp.figure(figsize = (width,height))
mp.grid(color='b', linestyle='-', linewidth=0.4)
x = df.Bowler
y = df.Ave
z = df.Econ
sns.lineplot(x,y,color="green")
mp.legend(['Average'])
mp.title("Bowling Average")
mp.xlabel("Bowlers")
mp.ylabel("Average")
```

```python
width = 22
height = 6
mp.figure(figsize = (width,height))
mp.grid(color='b', linestyle='-', linewidth=0.4)
sns.lineplot(x,z,color="red")
mp.title("Bowling Economy")
mp.legend(['Economy'])
mp.xlabel("Bowlers")
mp.ylabel("Economy")
```
