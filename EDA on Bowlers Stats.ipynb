# Exploratory Data Analysis on Bowlers Stats.

```python
import pandas as pd
import numpy as np

dt = pd.read_csv("E:/Data Sets/Bowler_data.csv")
df = pd.DataFrame(dt)
df.head(20)
```

```python
df = df.replace('-',np.nan)
```

```python
df.drop(columns = ['Unnamed: 0','Match_ID','Start Date','Player_ID','Ground'], axis = 1, inplace=True)
```

```python
df.dropna(subset = ['Overs'], axis = 0, inplace=True)
df
```

```python
df = df.replace(np.nan,0)
```

```python
df.dtypes
```

```python
df['Overs'] = df['Overs'].astype(float)
df['Mdns'] = df['Mdns'].astype(int)
df['Runs'] = df['Runs'].astype(int)
df['Wkts'] = df['Wkts'].astype(int)
df['Econ'] = df['Econ'].astype(float)
df['Ave'] = df['Ave'].astype(float)
df['SR'] = df['SR'].astype(float)
df['Opposition'] = df['Opposition'].astype('category')
df['Bowler'] = df['Bowler'].astype('category')
df.dtypes
```

```python
df['Opposition'] = df['Opposition'].str.lstrip('v.!?" " \n\t')
```

```python
df
```

# Analyzing the Top-Class Bowlers Stats!

```python
top = df[['Overs','Mdns','Runs','Wkts','Econ','Ave','Bowler']]
```

```python
top
```

```python
top = top.groupby(['Bowler']).sum()
top
```

```python
top = top.reset_index()
top
```

```python
top['Ave'] = top['Runs']/top['Wkts']
top
```

```python
top['Econ'] = top['Runs']/top['Overs']
top
```

```python
top.drop(columns = ['Overs'],inplace=True)
```

```python
top = top[top['Wkts'] >= 10.0]
top = top[top['Ave'] <= 25.0]
top
```

```python
top['Bowler'] = top['Bowler'].astype('str')
```

# Bowling Average of the world's Top-Class Bowlers.

```python
import matplotlib.pyplot as mp
import seaborn as sns
%matplotlib inline
```

```python
width = 22
height = 8
mp.figure(figsize = (width,height))
mp.grid(color='b', linestyle='-', linewidth=0.4)
sns.lineplot(top.Bowler,top.Ave,color="green",marker='o')
mp.legend(['Average'], fontsize=20)
mp.title("Bowling Average", fontsize=20)
mp.xlabel("Bowlers", fontsize=20)
mp.ylabel("Average", fontsize=20)
```

# Bowling Economy of the world's Top-Class Bowlers.

```python
width = 22
height = 8
mp.figure(figsize = (width,height))
mp.grid(color='b', linestyle='-', linewidth=0.4)
sns.lineplot(top.Bowler,top.Econ,color="red", marker='o')
mp.title("Bowling Economy", fontsize=20)
mp.legend(['Economy'], fontsize=20)
mp.xlabel("Bowlers", fontsize=20)
mp.ylabel("Economy", fontsize=20)
```

# Wickets taken by the world's top-class bowlers.

```python
width = 22
height = 10
mp.figure(figsize = (width,height))
mp.grid(color='b', linestyle='-', linewidth=0.1)
mp.bar(top.Bowler,top.Wkts,color="green")
mp.xlabel("Bowlers", fontsize=25)
mp.ylabel("Average", fontsize=25)
mp.title("Total Wickets Taken", fontsize=30)
```
